```python
import requests
import json
import pandas as pd
```

```python
url = "https://api-baseball.p.rapidapi.com/standings"

querystring = {"league":"1","season":"2020"}

headers = {
	"X-RapidAPI-Key": "55194652f4msh40bd2e5941627eap1b1b68jsn4f55d7d0851b",
	"X-RapidAPI-Host": "api-baseball.p.rapidapi.com"
}

response = requests.get(url, headers=headers, params=querystring)
```

```python
if response.status_code == 200:
  data = response.json()

  formatted_json = json.dumps(data, sort_keys=True, indent = 10)
  #print(formatted_json)

team_names = []
games_played = []
wins = []
lose = []
points_for = []
points_against = []


for team in data['response'][0]:
  team_names.append(team['team']['name'])
  games_played.append(team['games']['played'])
  wins.append(team['games']['win']['total'])
  lose.append(team['games']['lose']['total'])
  points_for.append(team['points']['for'])
  points_against.append(team['points']['against'])

  teams_data = {
    'Team Names':team_names,
    'Games Played':games_played,
    'Wins': wins,
    'Lose': lose,
    'For': points_for,
    'Against':points_against
}

df = pd.DataFrame(teams_data)
df
```
